```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset into a pandas DataFrame
dataset = pd.read_csv('Social Media Data for DSBA.csv')

# Visualize the relationships between numeric features using a pair plot
numeric_columns = ['Yearly_avg_view_on_travel_page', 'Yearly_avg_comment_on_travel_page',
                   'Daily_Avg_mins_spend_on_traveling_page', 'week_since_last_outstation_checkin',
                   'total_likes_on_outstation_checkin_given']

sns.pairplot(dataset[numeric_columns], diag_kind='kde')
plt.show()
```
